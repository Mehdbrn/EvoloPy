```python
import pandas as pd
import numpy as np


file_path ='/kaggle/input/2-attributes-hm1/12.xlsx'
df = pd.read_excel(file_path)

print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])
print(df)
```

```python
train_data_A = df.iloc[0:8, :]
train_data_B = df.iloc[10:18, :]

column1_mean_A = df.iloc[0:8, 0].mean()
column1_median_A = df.iloc[0:8, 0].median()
column2_mean_A = df.iloc[0:8, 1].mean()
column2_median_A = df.iloc[0:8, 1].median()

column1_mean_B = df.iloc[10:18, 0].mean()
column1_median_B = df.iloc[10:18, 0].median()
column2_mean_B = df.iloc[10:18, 1].mean()
column2_median_B = df.iloc[10:18, 1].median()

test_data = df.iloc[[9, 8, 18, 19], :]

print("Mean of Column 1_A:", column1_mean_A)
print("Median of Column 1_A:", column1_median_A)
print("Mean of Column 2_A:", column2_mean_A)
print("Median of Column 2_A:", column2_median_A)

print()

print("Mean of Column 1_B:", column1_mean_B)
print("Median of Column 1_B:", column1_median_B)
print("Mean of Column 2_B:", column2_mean_B)
print("Median of Column 2_B:", column2_median_B)

print()

print("Test Data:")
print(test_data)
print("train_data_A")
print(train_data_A)
print("train_data_B")
print(train_data_B)
```

```python
correct_predictions = 0
for i in range(len(test_data)):
    x = test_data.iloc[i, 0]
    y = test_data.iloc[i, 1]
    
    ac1 = abs(column1_mean_A - x)
    ac2 = abs(column2_mean_A - y)
    ac=ac1+ac2
    
    bc1= abs(column1_mean_B - x)
    bc2 = abs(column2_mean_B - y)
    bc=bc1+bc2
    
    print(ac)
    print()
    print(bc)
    print()
    if(bc>ac):
        predicted_class = "A"
        print("dade shomare"+ str(i+1) + "motalegh be class A hast")
    else:
        predicted_class = "B"
        print("dade shomare"+ str(i+1) + "motalegh be class B hast")
        
    print()
    
    true_class = test_data.iloc[i, 2]
    
    if predicted_class == true_class:
        correct_predictions += 1

accuracy = correct_predictions / len(test_data) * 100
print("Deghat:", accuracy, "%")
    
```
